# Error Case Mining

This notebook reads each model's `test_predictions.csv`, computes disease-level operating thresholds, and writes high-margin false-positive / false-negative cases.

Outputs per model:
- `false_positive.csv`
- `false_negative.csv`


```ipython
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT))

CHECKPOINT_DIR = PROJECT_ROOT / 'checkpoints'
MODELS = ['densenet', 'efficientnet', 'vit']
MAX_CASES = 12

print(f'Project root: {PROJECT_ROOT}')
print(f'Checkpoint dir: {CHECKPOINT_DIR}')
```

```ipython
from src.analysis.run_error_cases import analyze_model

all_outputs = {}
for model_key in MODELS:
    outputs = analyze_model(CHECKPOINT_DIR, model_key, MAX_CASES)
    all_outputs[model_key] = outputs
    print(
        f"{model_key}: FP={len(outputs['false_positive.csv'])}, "
        f"FN={len(outputs['false_negative.csv'])}"
    )
```

```ipython
for model_key in MODELS:
    print(f'\n[{model_key}] false_positive.csv')
    display(all_outputs[model_key]['false_positive.csv'])
    print(f'[{model_key}] false_negative.csv')
    display(all_outputs[model_key]['false_negative.csv'])
```
